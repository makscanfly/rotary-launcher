```python
import asyncio
from BLE_remote_control import LauncherRemoteControl
from datetime import datetime
import csv
from pathlib import Path
```

```python
# Inicjalizacja obiektu i łączenie z wyrzutnią
RemoteControl = LauncherRemoteControl()

await RemoteControl.connect()
print("Połączono:", RemoteControl.is_connected)
```

```python
# sprawdzenie statusu połączenia
print("połączono:", RemoteControl.is_connected)
```

```python
await RemoteControl.LedOff()
```

```python
await RemoteControl.ElectromagnetOn()
```

```python
await RemoteControl.ElectromagnetOff()
```

```python
pwm = 750
await RemoteControl.Accelerate_to(pwm)
```

```python
await RemoteControl.Deaccelerate()
```

```python
await RemoteControl.StopMotor()
```

```python
'''
Kody poleceń
enum class Command : uint8_t {
  LedOn = 1,
  LedOff = 2,   
  ResetSampleCollecting = 3,    // przygotowana wcześniej funkcja
  FinishSampleCollecting  = 4,  // g_samples_ready = true, g_capture_enabled = false
  SetMotorPwm = 5,              // ustawia wypełnienie PWM'a zgodne z wartością podaną przez BLE
  StopMotor = 6,                // ustawienie wypełnienia na 0 niezależnie od wartości przekazanej prez BLE
  ElectromagnetOn = 7,
  ElectromagnetOff = 8,
  DirHigh = 9,
  DirLow = 10,
  Deaccelerate = 11
};
'''
```

```python
# odczyt pomiarów do zmiennej
samples = await RemoteControl.read_samples()

print(f'Ilość próbek:{len(samples)}\n Dziesięć pierwszych próbek: {samples[:10]}')
```

```python
# zapis do pliku

out_dir = Path("../../collected_data")
out_dir.mkdir(parents=True, exist_ok=True)

ts = datetime.now().strftime("%Y_%m_%d__%H_%M_%S")
filename = out_dir / f"samples_pwm{pwm}_{ts}.csv"

with open(filename, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["idx", "sample_uint32"])
    for i, s in enumerate(samples):
        w.writerow([i, s])

print("Zapisano:", filename)
```

```python
# zakańczanie połączenia BLE
await RemoteControl.disconnect()
print("Połączono:", RemoteControl.is_connected)
```
